```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.pyplot import figure
import mpld3
import seaborn as sns

df = pd.read_csv('cleaned_final_df.csv')
df.head(10)
```

```python
df.corr()
```

```python
df[['Ex-Showroom_Price','Displacement','Fuel_Tank_Capacity','City_Mileage']].corr()
```

```python

displacement_price, ax2 = plt.subplots()
ax2 = sns.regplot(x="Displacement", y="Ex-Showroom_Price", data=df)
displacement_price.subplots_adjust(left=0.2)

plt.ylim(0,)
plt.title("Variation of Price with Displacement")
mpld3.save_html(displacement_price,'displacement_price.html')
```

```python
mileage_price, ax3 = plt.subplots()
ax3 = sns.regplot(x="City_Mileage", y="Ex-Showroom_Price", data=df)
mileage_price.subplots_adjust(left=0.2)
plt.title("Variation of Price with Mileage")
plt.setp(ax3.get_xticklabels(), rotation=90)
plt.ylim(0,)
mpld3.save_html(mileage_price,'mileage_price.html')
```

```python
bodytype_price, ax4 = plt.subplots()
ax4 = sns.boxplot(x="Body_Type", y="Ex-Showroom_Price", data=df)
#plt.xticks(rotation=90)
plt.title("Variation of Price with Body Type")
bodytype_price.subplots_adjust(left=0.2)
plt.setp(ax4.get_xticklabels(), rotation=90, horizontalalignment='right')
mpld3.save_html(bodytype_price,'bodytype_price.html')
```

```python
gears_price, ax5 = plt.subplots()
ax5 = sns.boxplot(x="Gears", y="Ex-Showroom_Price", data=df)
#plt.xticks(rotation=90)
plt.title("Variation of Price with Number of Gears")
gears_price.subplots_adjust(left=0.2)
mpld3.save_html(gears_price,'gears_price.html')
```

```python
#grouping
df['Type'].unique()
```

```python
df_type = df[['Type','Body_Type','Ex-Showroom_Price']]
```

```python
# grouping results
df_type = df_type.groupby(['Type'],as_index=False).mean()
df_type
```

```python
df_type_body = df[['Type','Body_Type','Ex-Showroom_Price']]
type_body = df_type_body.groupby(['Type','Body_Type'],as_index=False).mean()
type_body
```

```python
group_pivot = type_body.pivot(index='Type',columns='Body_Type')
```

```python
group_pivot = group_pivot.fillna(0)
group_pivot
```

```python
#heat map to visualize realtion Body Type vs Price
plt.pcolor(group_pivot, cmap='GnBu')
plt.colorbar()
plt.show()
```

```python
bodytype_type_heatmap, ax = plt.subplots()
im = ax.pcolor(group_pivot, cmap='GnBu')

#label names
row_labels = group_pivot.columns.levels[1]
col_labels = group_pivot.index

#move ticks and labels to the center
ax.set_xticks(np.arange(group_pivot.shape[1]) + 0.5, minor=False)
ax.set_yticks(np.arange(group_pivot.shape[0]) + 0.5, minor=False)

#insert labels
ax.set_xticklabels(row_labels, minor=False)
ax.set_yticklabels(col_labels, minor=False)

#rotate label if too long
plt.xticks(rotation=90)

bodytype_type_heatmap.colorbar(im)
#plt.show()

mpld3.save_html(bodytype_type_heatmap,'bodytype_type_heatmap.html')
```
