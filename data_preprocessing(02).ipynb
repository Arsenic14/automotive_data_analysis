```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#reading the cleaned data set(cleaned and preprocessed using pandas and excel)
df = pd.read_csv('cleaned_cars_engage_2022.csv')
```

```python
df.head(10)
```

```python
df.dtypes
```

```python
avg_price=df['Ex-Showroom_Price'].astype('float').mean(axis=0)
print("Average price:", avg_price)
```

```python
df['Ex-Showroom_Price'].replace(np.nan, avg_price, inplace=True)
```

```python
df.head(10)
```

```python
df.describe()
```

```python
missing_data = df.isnull()
missing_data.head(10)
```

```python
#proper data type
df[["Ex-Showroom_Price"]] = df[["Ex-Showroom_Price"]].astype("float")
```

```python
# normalization
df["Length"] = df["Length"]/df["Length"].max()
df["Width"] = df["Width"]/df["Width"].max()
df['Height'] = df['Height']/df['Height'].max() 
df[["Length","Width","Height"]].head(10)
```

```python
df["Displacement"] = df["Displacement"].astype(int,copy=True)
```

```python
#Binning
%matplotlib inline
plt.hist(df["Displacement"])
plt.xlabel("Displacement")
plt.ylabel("count")
plt.title("DISPLACEMENT BINS")
```

```python
# dummy variable
dummy_fuel_type = pd.get_dummies(df["Fuel_Type"])
dummy_fuel_type.head(10)
```

```python
# merge "df" and "dummy_fuel_type" 
df = pd.concat([df, dummy_fuel_type], axis=1)

# drop "Fuel_Type" from "df"
df.drop("Fuel_Type", axis = 1, inplace=True)
```

```python
df.head(10)
```

```python
df.to_csv('cleaned_final_df.csv')
```
